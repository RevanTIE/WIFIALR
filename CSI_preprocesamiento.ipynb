## Debe aplicarse antes imputación de datos

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Funcion de Normalización de Datos
def normalization(df, col_headers):
    scaler = MinMaxScaler()
    scaled_df = scaler.fit_transform(df)
    scaled_df = pd.DataFrame(scaled_df, col_headers)
    return scaled_df
```
